```python
import matplotlib.pyplot as plt
import numpy as np
```

For our plots we will be using numbers from 1 to 10 as paramters on x axis. For y-axis will use Square root, Log and Log10 of these numbers. These can be easily calculated using numpy as shown below.

```python
numbers = np.arange(1, 11)
sqrts = np.sqrt(numbers)
logs = np.log(numbers)
log10s = np.log10(numbers)
```

Creating a plot using `matplotlib` as easy as calling `plot` function with paramters for x and y axis. Following small piece of code can plot these three for us very easily.

```python
plt.plot(numbers, logs)
plt.plot(numbers, sqrts)
plt.plot(numbers, log10s)

plt.show()
```

It looks pretty ok, but doesn't provide us much information like which lines belongs to which function and there are no labels. So, let's see how can we convert this to the following more meaningful and beautyful plot very easily using power of `matplotlib`

```python
plt.figure(figsize=(6, 4), dpi=200, facecolor='#F0F0F0')

plt.plot(numbers, logs, label='Log', color='dodgerblue', marker='D', ms=5, markerfacecolor='orange', markeredgecolor='green', markeredgewidth=0.5)
plt.plot(numbers, sqrts, label='Square Root', color='darkorange', marker='.', ls='--', lw=1, markerfacecolor='green', markeredgewidth=0)
plt.plot(numbers, log10s, label='Log10', color='seagreen', marker='o', ms=5, markeredgewidth=1, markerfacecolor='white', markeredgecolor='seagreen', lw=1)

plt.legend(loc=(1.1, 0.5))

xticks_labels = ['One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten']



ax = plt.gca()
ax.set_facecolor('#F1FFF1')

font = {'family': 'serif',
        'color':  'dodgerblue',
        'weight': 'normal',
        'size': 6,
        }
plt.xticks(range(1,11), xticks_labels, rotation=45, fontdict=font)
plt.xlabel(xlabel="Numbers 1 to 10", loc='right', fontdict=font)
plt.ylabel("Result", loc='bottom')
plt.title("Mathematial Functions!", loc="left")

plt.show()
```
